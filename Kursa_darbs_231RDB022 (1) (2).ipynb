Tāpēc es izmantoju TKinker bibliotēku, lai īstenotu garfisko interfeisu, kurš sastāvēja no canvas objekta, kur tika zīmēta līkne, kā arī aplis, kurš kustās pa līkni nospiežot pogu. Kā arī zem canvas objekta ir izveidots label objekts, kurš attēlo informāciju lietotājam, ka viņš var izvēlēties 5 kontrolpunktus nospiežot kreiso peles taustiņu uz Canvas objekta. Turklāt ar Tkinker palīdzību tiek attēlota poga "Animēt", kura izsauc apļa zīmēšanas un animēšanas metodes.

Turklāt ar Tkinter bibliotēkas metodi bind uz canvas noklikšķinot es iegūstu x un y koordinātes, kas ir kontrolpunktu koordinātes, kuras es pēc tam pievienoju sarakstā, lai tālāk izmantotu tos Bezier algoritmā līknes zīmēšanai, objekta zīmēšanai un objekta animācijai. Ar šīm koordinātēm es arī metodes kontrolpunktu_pievienosana iekšienē uzzīmēju kontrolpunktus ar Tkinter bibliotēkas metodi create_oval. Pēc tam tiek izsaukta metode zimet_likni, ja lietotājs ir noklikšķinājis 5 kontrolpunktus. 

Metodē zimet_likni tiek realizēts Bezjē līknes veidošanas algoritms, kur x un y līknes koordinātes tiek saglabātas sarakstā liknes_punkti. Metodē tiek realizēta arī līknes zīmēšana uz canvas objekta no saraksta liknes_punkti vērtībām, izmantojot create_line metodi, kas ir arī Tkinker bibliotēkas metode.

Tad, kad ir nospiesta poga "Animēt", tiek izsaukta metode, kas izsauc divas metodes, ja ir pievienoti pieci kontrolpunkti. Pirmā metode, kas tiek izsaukta ir zimet_objektu, kas tiek definēta objekts mainīgajā, kas tālāk tiks izmantots otrajā metodē. Objekts mainīgajam tiek saglabāts pirmās metodes rezūltāts, kas ir create_oval metodes rezultātā iegūts aplis melnā krāsā pirmajā kontrolpunktā. Tad tiek izsaukta otrā metode objekta_animacija, kas sevī ietver sevī Bezjē līknes veidošanas algoritmu, lai pārvietotu objektu pa šo līkni, tad tiek izaukta Tkinter bibliotēkas metode cords, kur tiek padots objekts mainīgais un šīs te līknes koordinātes, kas nodrošina šo te apļa koordināšu maiņu, tad tiek izsaukta metode update, kas atjauno canvas logu, lai šīs koordinātu izmaiņas attēlotos Canvas logā un tad tiek realizēta pauze starp kadriem.

### Kursa darba praktiskā daļa

```python
# Importēju Tkinter bibliotēku
import tkinter as tk

# Izveidoju klasi "App"
class App:
    # Konstruktorā inicializē sākotnējos objektus
    def __init__(self, master):
        # Norādu uz vecāko loga objektu
        self.master = master
        # Izveidoju galvenā loga virsrakstu
        self.master.title("Bezjē Algoritma attēlošana")
        
        # Izveidoju Canvas objektu, kurā tiks zīmēta līkne un aplis, kurš pārvietojas pa līkni
        self.canvas = tk.Canvas(master, width=600, height=400, background='white')
        self.canvas.pack() # Attēloju Canvas objektu

        # Saraksts, kurā glabāsies kontrolpunkti
        self.punkti = []
        # Definēju mainīgo objekta identifikatoru
        self.objekts = None
        # Animācijas ātrums (milisekundēs)
        self.animacijas_atrums = 1

        # Tiek realizēta notikuma piesaistīšana(kreisais taustiņš uz Canvas)
        self.canvas.bind("<Button-1>", self.kontrolpunktu_pievienosana)

        # Pievienoju informāciju lietotājam cik kontrolpunktus var pievienot
        self.label = tk.Label(master, text="Noklikšķiniet, lai pievienotu kontrolpunktu (pieļaujami tikai 5)")
        self.label.pack()
        # Izveidoju pogu, kas ir atbildīga par animācijas veidošanu
        self.button = tk.Button(master, text="Animēt", command=self.zimet)
        self.button.pack()

    # Metode, kas pievieno kontrolpunktu sarakstam, atšķirībā no tā kur lietotājs noklikšķināja
    def kontrolpunktu_pievienosana(self, event):
        if len(self.punkti) < 5:  # Pārbaudu, vai nav pievienoti vairāk kā 5 kontrolpunkti
            # Iegūst peles klikšķa koordinātas
            x = event.x 
            y = event.y
            self.punkti.append((x, y))  # Pievieno kontrolpunktu sarakstam
            # Zīmēju kontrolpunktus uz Canvas ar create_oval metodi
            self.canvas.create_oval(x - 5, y - 5, x + 5, y + 5, fill='red')
           
            # Ja ir pievienoti 5 kontrolpunkti, tad izsauc metodi, lai zīmētu likni
            if len(self.punkti) == 5:
                self.zimet_likni(self.punkti)

    # Izveidoju metodi, kas izsauc objekta zīmēšanas metodes
    def zimet(self):
        if len(self.punkti) == 5:  # Pārbaudu, vai ir pievienoti 5 kontrolpunkti
            self.objekts = self.zimet_objektu()  # Izsaucu metodi, lai zīmētu kustīgo objektu
            self.objekta_animacija(self.punkti)  # Izsauc metodi, lai animētu objektu pa līkni

    # Metode, kas zīmē Bezier līkni, izmantojot kontrolpunktu sarakstu
    def zimet_likni(self, punkti):
        liknes_punkti = []  # Saraksts, kurā glabāsies Bezier līknes punkti
        t = 0 # parametrs t, kas pieder intervālam no 0 līdz 1
        ts = 0.00001 # solis, ar kuru mainās parametrs t
        while t <= 1.0:  # kamēr nav sasniegts interavala gala punkts
            
            B0 = 1*(1-t)**4 #rēķinam Beršteina 0.polinomu
            B1 = 4* (1-t)**3 * t #rēķinam Beršteina 1.polinomu
            B2 = 6* (1-t)**2 * t**2 #rēķinam Beršteina 2.polinomu
            B3 = 4* (1-t)**1 * t**3 #rēķinam Beršteina 3.polinomu
            B4 = 1*t**4 #rēķinam Beršteina 4.polinomu
            
            #rēķinam līknes pikseļa koordinātes pēc x ass
            x = int(punkti[0][0]*B0 + punkti[1][0]*B1 + punkti[2][0]*B2 + punkti[3][0]*B3 + punkti[4][0]*B4)
            #rēķinam līknes pikseļa koordinātes pēc y ass
            y = int(punkti[0][1]*B0 + punkti[1][1]*B1 + punkti[2][1]*B2 + punkti[3][1]*B3 + punkti[4][1]*B4)
            liknes_punkti.append((x, y)) #pievienojam koordinātes sarakstam
            t+=ts #palielinam t vērtību

        return self.canvas.create_line(liknes_punkti, fill='blue')  # Zīmē Bezier līkni uz Canvas

    # Metode, kas zīmē kustīgu objektu uz sākotnējā kontrolpunkta
    def zimet_objektu(self):
        return self.canvas.create_oval(self.punkti[0][0] - 10, self.punkti[0][1] - 10, self.punkti[0][0] + 10, self.punkti[0][1] + 10, fill='black')

    # Metode, kas animē objektu pa Bezier līkni
    def objekta_animacija(self,punkti):
        if self.objekts:
            t = 0 # parametrs t, kas pieder intervālam no 0 līdz 1
            ts = 0.005 # solis, ar kuru mainās parametrs t
            while t <= 1.0:  # kamēr nav sasniegts interavala gala punkts
                B0 = 1*(1-t)**4 #rēķinam Beršteina 0.polinomu
                B1 = 4* (1-t)**3 * t #rēķinam Beršteina 1.polinomu
                B2 = 6* (1-t)**2 * t**2 #rēķinam Beršteina 2.polinomu
                B3 = 4* (1-t)**1 * t**3 #rēķinam Beršteina 3.polinomu
                B4 = 1*t**4 #rēķinam Beršteina 4.polinomu
                #rēķinam līknes pikseļa koordinātes pēc x ass
                x = int(punkti[0][0]*B0 + punkti[1][0]*B1 + punkti[2][0]*B2 + punkti[3][0]*B3 + punkti[4][0]*B4)
                #rēķinam līknes pikseļa koordinātes pēc y ass
                y = int(punkti[0][1]*B0 + punkti[1][1]*B1 + punkti[2][1]*B2 + punkti[3][1]*B3 + punkti[4][1]*B4)
                t+=ts
                self.canvas.coords(self.objekts, x - 10, y - 10, x + 10, y + 10)  # Pārvieto objektu pa līkni
                self.canvas.update() #Atjainina Canvas logu
                self.master.after(self.animacijas_atrums) # Pauze starp kadriem animācijā

root = tk.Tk() #Izveidoju galveno loga objektu
app = App(root) #Izveidoju app objektu, izmantojot iepriekš definēto klasi. 
root.mainloop() #Izveidoju grafiskā interfeisa ciklu
```
